3. Use the following function to download the example data for the course:

```python
def download_data(force=False):
    """Download and extract course data from Zenodo."""
    import urllib.request, zipfile, os
    
    zip_path = 'data.zip'
    data_dir = 'data'
    
    if not os.path.exists(zip_path) or force:
        print("Downloading course data")
        urllib.request.urlretrieve(
            'https://zenodo.org/records/16954427/files/data.zip?download=1',
            zip_path
        )
        print("Download complete")
    else:
        print("Download file already exists")
        
    if not os.path.exists(data_dir) or force:
        print("Extracting data files...")
        with zipfile.ZipFile(zip_path, 'r') as zip_ref:
            zip_ref.extractall(data_dir)
        print("Data extracted")
    else:
        print("Data directory already exists")

download_data()
```

A primary question is whether house sale prices can be accurately predicted based on structural characteristics (e.g., size, quality rating) and neighborhood. A prediction tool, such as a regression model or machine learning price estimator, could help forecast property values. Stakeholders for this tool would include prospective home buyers, real estate agents, mortgage lenders, and local governments.

However, there are important practical and ethical considerations. Predictive models may unintentionally reinforce housing inequality if they systematically undervalue properties in certain neighborhoods. For example, lenders using such models for loan approvals could contribute to biased lending practices. Additionally, real estate agents or sellers could use the tool in ways that inflate prices unfairly. Transparency in how the model works and safeguards against neighborhood-based bias would be critical.

Other valuable data that could improve predictions include current mortgage interest rates, household income levels in each neighborhood, local property tax rates, and details about recent renovations or upgrades. These additional variables would allow for more realistic and fair modeling of housing prices.
